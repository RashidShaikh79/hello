# Missing Indicator Imputation

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv('train (5).csv', usecols=['Age', 'Fare', 'Survived'])
df.head()
```

## Create X & Y

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

## Train & Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

## Review Without “Missing Indicator Method” Technique

```python
si = SimpleImputer()
x_train_trf = si.fit_transform(x_train)
x_test_trf = si.fit_transform(x_test)
```

```python
x_train_trf
```

```python
x_test_trf
```

## Call Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()

clf.fit(x_train_trf,y_train)

y_pred = clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

## Define Missing Indicator

```python
mi = MissingIndicator()
mi.fit(x_train)
```

```python
MissingIndicator()
```

```python
mi.features_
```

```python
x_train_missing = mi.transform(x_train)

x_train_missing
```

## Transform: Train Missing

```python
x_test_missing = mi.transform(x_test)

x_test_missing
```

## Transform: Test Missing

```python
x_train['Age_NA'] = x_train_missing

x_test
```

## Create New Column

```python
x_test['Age_NA'] = x_test_missing

x_train
```

## Writing Code again and Check Accuracy

```python
si = SimpleImputer()

x_train_trf2 = si.fit_transform(x_train)
x_test_trf2 = si.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(x_train_trf2, y_train)
y_pred = clf.predict(x_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
