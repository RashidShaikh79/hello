### Fetching Data Using Web Scraping.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

## use "soup" for BeautifulSoup

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
# print(soup.prettify())
```

### Extract the heading line

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

### Extract company rating

```python
for k in soup.find_all('span', class_='companyCardWrapper__companyRatingValue'):
    print(k.text.strip())
```

```python
for k in soup.find_all('span'):
    print(k.text.strip())
```

### Extract Card Data

```python
for j in soup.find_all('span', class_='companyCardWrapper__interLinking'):
    print(j.text.strip())
```
