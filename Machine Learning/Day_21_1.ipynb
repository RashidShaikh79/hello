```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train (5).csv')
df.head()
```

```python
df = pd.read_csv('train (5).csv')[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
df.head()
```

# Drop NA Value

```python
df.dropna(inplace=True)
df.sample(5)
```

# Separate X ana Y

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

# Check Accuracy for Logistics Regression

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

# Applying Feature Construction

## Create New Column

```python
x['Family_size'] = x['SibSp'] + x['Parch'] + 1
```

```python
x.head()
```

# Apply New Function

```python
def myfunc(num):
    if num == 1:
        # alone
        return 0
    elif num > 1 and num <= 4:
        # Small family
        return 1
    else:
        # Large family
        return 2
```

```python
myfunc(4)
```

# Apply M Function

```python
x['Family_type'] = x['Family_size'].apply(myfunc)
x.head()
```

# Drop Unwanted Columns

```python
x.drop(columns=['SibSp', 'Parch', 'Family_size'],inplace=True)
```

```python
x.head()
```

# Review Accuracy After Feature Contruction

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```
